# EDA (Exploratory Data Analysis)

```python
start_date = '2010-01-01'
end_date = '2023-6-19'
symbol = 'AAPL'
data = yf.download(symbol, start=start_date, end=end_date)
```

## Feature Engineering

```python
# Calculate Price Change 
#Capture the daily volatility of the stock
data['Price Change'] = data['Close'].diff()
data['Price Change'].fillna(0, inplace=True)

# Calculate Price Range 
#daily price range  insights into the intraday price movement and volatility

data['Price Range'] = data['High'] - data['Low']

#Calcuate 30 day moving average 
#Moving averages can help identify trends and smoothen out short-term fluctuations.

data['MA_30'] = data['Close'].rolling(window=30).mean()
data['MA_30'].fillna(0, inplace=True)

#RSI(Relative Strength Index)
#momentum oscillator that measures the speed and change of price movements. It can help identify overbought or oversold conditions.
#RSI score of 70 typically indicating overbought conditions and values below 30 indicating oversold conditions.

delta = data['Close'].diff()
gain = delta.mask(delta < 0, 0)
loss = -delta.mask(delta > 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data['RSI'] = 100 - (100 / (1 + rs))
data['RSI'].fillna(0, inplace=True)

# Calculate Volume Changes
data['Volume Change'] = data['Volume'].pct_change()
data['Volume Change'].fillna(0, inplace=True)

# Calculate MACD
#By calculating the MACD, you can gain insights into the strength and direction of the price trends in the underlying asset. 
#Positive MACD values suggest bullish signals, while negative MACD values indicate bearish signals. 
data['EMA_12'] = data['Close'].ewm(span=12).mean()
data['EMA_26'] = data['Close'].ewm(span=26).mean()
data['MACD'] = data['EMA_12'] - data['EMA_26']

# Extract Day of the Week
data['Day of Week'] = data.index.dayofweek
```

```python
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error

# Step 1: Split data into features (X) and target variable (y)
X = data.drop('Open', axis=1)
y = data['Open']

# Step 2: Define the number of folds for cross-validation
k = 5

# Step 3: Instantiate your decision tree model
model = DecisionTreeRegressor()

# Step 4: Create a cross-validation object
cv = KFold(n_splits=k)

# Step 5: Perform feature selection and evaluation
evaluation_scores = []
for train_index, test_index in cv.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Instantiate feature selector
    feature_selector = SelectKBest()

    # Fit feature selector on training data
    feature_selector.fit(X_train, y_train)

    # Transform the training and test data to selected features
    X_train_selected = feature_selector.transform(X_train)
    X_test_selected = feature_selector.transform(X_test)

    # Train a decision tree model on the selected features
    model.fit(X_train_selected, y_train)

    # Evaluate the model on the test data
    y_pred = model.predict(X_test_selected)
    score = np.sqrt(mean_squared_error(y_test, y_pred))
    evaluation_scores.append(score)

# Step 6: Calculate the average performance
average_score = sum(evaluation_scores) / len(evaluation_scores)
print("Average Performance:", average_score)

# Step 7: Determine the best subset of features
best_features = feature_selector.get_support(indices=True)
print("Best Features:", X.columns[best_features])

# Step 8: Retrain decision tree model using the best subset of features on the entire dataset
X_selected = feature_selector.transform(X)
model.fit(X_selected, y)

# Step 9: Make predictions or perform further analysis
# Use the selected features (X_selected) and the trained decision tree model to make predictions or perform analysis
```

# Conclusion

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Extract the actual values and predicted values for each predictor
actual_values = data['Open']
dt_prediction = data['dt_prediction_open']
kcross_dt_prediction = data['kcross_dt_prediction_open']
random_forest_regression = data['Random_Forest_Regressor_Pred']

# Calculate RMSE for each predictor
dt_rmse = np.sqrt(mean_squared_error(actual_values, dt_prediction))
kcross_dt_rmse = np.sqrt(mean_squared_error(actual_values, kcross_dt_prediction))
rf_regression_rmse = np.sqrt(mean_squared_error(actual_values, random_forest_regression))

# Print the RMSE values
print("Decision Tree RMSE:", dt_rmse)
print("K-Cross Decision Tree RMSE:", kcross_dt_rmse)
print("Random Forest Regression RMSE:", rf_regression_rmse)
```

Decision Tree: The Decision Tree model achieved an RMSE of 0.3573. This indicates that, on average, the predicted values deviated from the actual values by approximately 0.3573 units. While the Decision Tree model provides a reasonable prediction, there is room for improvement.

K-Cross Decision Tree: The K-Cross Decision Tree model performed better with an RMSE of 0.1487. The lower RMSE suggests that the K-Cross Decision Tree model produced more accurate predictions compared to the regular Decision Tree model. It benefited from the feature selection process using k-fold cross-validation, resulting in improved performance.

Random Forest Regression: The Random Forest Regression model yielded the lowest RMSE of 0.1231. This implies that the Random Forest model produced the most accurate predictions among the three models. Random Forests are known for their ability to capture complex relationships and handle high-dimensional data, making them well-suited for stock prediction tasks.

In conclusion, the Random Forest Regression model outperformed both the Decision Tree and K-Cross Decision Tree models in terms of predictive accuracy. It achieved the lowest RMSE, indicating a better fit to the actual stock values. Therefore, the Random Forest Regression model is recommended for stock prediction based on the given results.
